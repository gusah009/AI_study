```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

train = pd.read_csv("train.csv")
train = train.dropna(subset=['Sex', 'Age', 'Fare'])

testing = pd.read_csv("test.csv")
```

```python
# Setting
sex_weight : 1.1
Pclass_weight : 1
age_weight : 1
fare_weight : 1
cnt_similar : 3

""" ======================= 여기 오래 걸림 ======================= """

for i in range(10):
    sex_weight += 0.1
    is_survive = []
    for test_index, test_data in testing.iterrows():
        similarity_similarity_scores = []
        similarity_scores = []
        for train_index, train_data in train.iterrows():
            similarity_score = 0
            if test_data["Sex"] == train_data["Sex"]:
                similarity_score += sex_weight

            # Pclass가 같으면 같을수록 값이 커짐
            if math.isnan(test_data["Pclass"]) != True:
                maximum = testing["Pclass"].max() - testing["Pclass"].min()
                similarity_score += (maximum - abs(test_data["Pclass"] - train_data["Pclass"]))/maximum * Pclass_weight

            # 나이가 같으면 같을수록 값이 커짐
            if math.isnan(test_data["Age"]) != True:
                maximum = testing["Age"].max() - testing["Age"].min()
                similarity_score += (maximum - abs(test_data["Age"] - train_data["Age"]))/maximum * age_weight

            # 표 가격이 같으면 같을수록 값이 커짐
            if math.isnan(test_data["Fare"]) != True:
                maximum = testing["Fare"].max() - testing["Fare"].min()
                similarity_score += (maximum - abs(test_data["Fare"] - train_data["Fare"]))/maximum * fare_weight

            similarity_scores.append(similarity_score)

        similarity_scores = np.array(similarity_scores)
        # 비효율적이긴함
        survive_score = 0
        for i in range(cnt_similar):
            max_index = similarity_scores.argmax()
            survive_score += train.iloc[max_index]["Survived"]
            similarity_scores = np.delete(similarity_scores, max_index)

        if survive_score > int(cnt_similar/2):
            is_survive.append(1)
        else:
            is_survive.append(0)
    submission = pd.DataFrame({'PassengerId':testing['PassengerId'],'Survived':is_survive})
    submission.to_csv('answer' + str(i+1) + '.csv', index=False)

""" ======================= 여기 오래 걸림 ======================= """

# solution = pd.read_csv("gender_submission.csv")
# rate = 0
# for solution_index, solution_data in solution.iterrows():
#     if solution_data["Survived"] == is_survive[solution_index]:
#         rate += 1
# print("{0:.2f}%".format(rate / len(solution) * 100))
# # 69.38%
```
