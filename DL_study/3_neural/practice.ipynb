### 계단함수 그래프

```python
x = np.arange(-5.0, 5.0, 0.1)
y = af.step_function(x)
plt.plot(x, y)
plt.show()
```

### 계단함수 그래프

```python
x = np.arange(-5.0, 5.0, 0.1)
y = af.sigmoid(x)
plt.plot(x, y)
plt.show()
```

### 계단함수, sigmoid 비교

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = af.sigmoid(x)
y2 = af.step_function(x)
plt.plot(x, y1, x, y2, '--')
plt.show()
```

### ReLu 그래프 그리기

```python
x = np.arange(-4.0, 4.0, 0.1)
y = af.relu(x)
plt.plot(x, y)
plt.show()
```
