```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
def Categorization(x):
    category_age = 10
    if pd.notna(x):
        return int(x // category_age * category_age)
    return 9999
```

```python
train_age_category = train.copy()
train_age_category['Age'] = train['Age'].apply(Categorization)
chartData = train_age_category['Age'].value_counts().sort_index()
chartData
```

```python
chartData.drop(index=9999)
```

```python
chartData.drop(index=9999, inplace=True)
plt.plot(chartData)
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show()
```
